<h1 style="color:DodgerBlue">Создание экземпляров классов (объектов) и взаимодействие объектов между собой</h1>

### Создание экземпляров классов (объектов)

В C#, экземпляр класса создается с помощью оператора `new`, который выделяет память для нового объекта и вызывает конструктор класса для инициализации этого объекта. Конструктор — это специальный метод в классе, который имеет то же имя, что и класс, и используется для установки начального состояния объекта.

#### Пример создания экземпляров классов Vehicle и Car

```csharp
Vehicle myVehicle = new Vehicle();
Car myCar = new Car();
```

В этом примере создаются два объекта: один для класса `Vehicle` и другой для класса `Car`. Конструкторы этих классов будут автоматически вызваны при создании объектов.

### Взаимодействие объектов между Собой

Взаимодействие объектов между собой означает, что один объект может вызывать методы другого объекта или обращаться к его свойствам (если они доступны). Это позволяет объектам обмениваться информацией и координировать свои действия.

#### Пример взаимодействия объектов

```csharp
myVehicle.Color = "Red";
myVehicle.Speed = 60;

myCar.Color = "Blue";
myCar.FuelLevel = 100;
myCar.Refuel(50); // Увеличиваем уровень топлива на 50
```

В этом примере мы устанавливаем свойства `Color` и `Speed` для объекта `myVehicle`, а также свойства `Color` и `FuelLevel` для объекта `myCar`. Затем мы вызываем метод `Refuel` у объекта `myCar`, чтобы увеличить его уровень топлива.

### Закрепление правил использования модификаторов доступа

Модификаторы доступа определяют уровень доступности членов класса из других частей кода. Важно правильно использовать модификаторы доступа для обеспечения безопасности и инкапсуляции данных.

#### Примеры модификаторов доступа в классах Vehicle и Car

```csharp
public class Vehicle
{
    public string Color { get; set; } // public свойство
    private int speed; // private поле

    protected void SetSpeed(int value) // protected метод
    {
        speed = value;
    }
}

public class Car : Vehicle
{
    private int fuelLevel; // private поле

    public void Refuel(int amount) // public метод
    {
        fuelLevel += amount;
    }

    protected void CheckSpeedLimit() // protected метод
    {
        if (speed > 120)
        {
            Console.WriteLine("Превышен скоростной режим!");
        }
    }
}
```

В этом примере:
- Свойство `Color` в классе `Vehicle` объявлено как `public`, что позволяет доступ к нему из любого места в коде.
- Поле `speed` в классе `Vehicle` объявлено как `private`, что ограничивает доступ к нему только внутри класса `Vehicle`.
- Метод `SetSpeed` в классе `Vehicle` объявлен как `protected`, что позволяет доступ к нему из класса `Vehicle` и его производных классов.
- Поле `fuelLevel` в классе `Car` объявлено как `private`, ограничивая доступ к нему только внутри класса `Car`.
- Методы `Refuel` и `CheckSpeedLimit` в классе `Car` демонстрируют использование модификаторов доступа для контроля над тем, как объекты могут взаимодействовать друг с другом.

### Заключение

Создание экземпляров классов и взаимодействие между объектами являются ключевыми аспектами объектно-ориентированного программирования в C#. Правильное использование модификаторов доступа обеспечивает безопасность данных и инкапсуляцию, позволяя разработчикам создавать более надежные и управляемые программы.

<h4 style="color:DodgerBlue">Для проверки напишите пример кода на основе классов Car и Vehicle ниже в блоке:</h4>

----

````
public class Vehicle
{
    public string Color { get; set; } // public свойство
    public int speed; // private поле

    protected void SetSpeed(int value) // protected метод
    {
        speed = value;
    }
}

public class Car : Vehicle
{
    public int fuelLevel; // private поле

    public void Refuel(int amount) // public метод
    {
        fuelLevel += amount;
    }

    protected void CheckSpeedLimit() // protected метод
    {
        if (speed > 120)
        {
            Console.WriteLine("Превышен скоростной режим!");
        }
    }
}

Vehicle myVehicle = new Vehicle();
myVehicle.Color = "Red";
myVehicle.speed = 60;

Car myCar= new Car();
myCar.Color = "Blue";
myCar.fuelLevel = 100;
myCar.Refuel(50); // Увеличиваем уровень топлива на 50
````

Ниже в блоке по примеру создайте базовый класс Animal и производные классы (2-3 например Dog, Cat и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также статические и обычными атрибуты и метода класса с реализацией взаимодействия между объектами.

````
public class Animal
{
    // Публичное свойство (доступно отовсюду)
    public string Name { get; set; }
    
    // Публичное свойство для цвета
    public string Color { get; set; }
    
    // Приватное поле (инкапсуляция)
    private int age;
    
    // Приватное поле для уровня голода
    private int hungerLevel;
    
    // Приватное поле для уровня энергии
    private int energyLevel;
    
    // Конструктор
    public Animal()
    {
        hungerLevel = 50;
        energyLevel = 100;
    }
    
    // Публичное свойство для доступа к возрасту с валидацией
    public int Age
    {
        get { return age; }
        set 
        { 
            if (value >= 0)
                age = value;
            else
                Console.WriteLine("Возраст не может быть отрицательным!");
        }
    }
    
    // Защищенный метод для установки возраста (аналог SetSpeed)
    public void SetAge(int value)
    {
        if (value >= 0)
            age = value;
        else
            Console.WriteLine("Возраст не может быть отрицательным!");
    }
    
    // Защищенный метод для установки уровня голода
    protected void SetHunger(int value)
    {
        hungerLevel = Math.Clamp(value, 0, 100);
    }
    
    // Защищенный метод для установки уровня энергии
    protected void SetEnergy(int value)
    {
        energyLevel = Math.Clamp(value, 0, 100);
    }
    
    // Публичные методы для получения значений
    public int GetHungerLevel()
    {
        return hungerLevel;
    }
    
    public int GetEnergyLevel()
    {
        return energyLevel;
    }
    
    // Виртуальный метод для издания звука
    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издает звук");
    }
    
    // Метод для вывода информации
    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Имя: {Name}, Цвет: {Color}, Возраст: {age}, Голод: {hungerLevel}, Энергия: {energyLevel}");
    }
}

// Производный класс Dog (аналог Car)
public class Dog : Animal
{
    // Публичное поле для породы (в примере fuelLevel был public)
    public string breed;
    
    // Приватное поле для уровня счастья
    private int happinessLevel;
    
    // Конструктор
    public Dog() : base()
    {
        happinessLevel = 50;
    }
    
    // Публичный метод для кормления (аналог Refuel)
    public void Feed(int amount)
    {
        Console.WriteLine($"{Name} ест {amount} порций корма");
        SetHunger(GetHungerLevel() - amount);
        happinessLevel += 10;
        Console.WriteLine($"Голод: {GetHungerLevel()}, Счастье: {happinessLevel}");
    }
    
    // Защищенный метод для проверки состояния (аналог CheckSpeedLimit)
    protected void CheckHealthStatus()
    {
        if (GetHungerLevel() > 80)
        {
            Console.WriteLine($"{Name} очень голоден! Нужно покормить!");
        }
        else if (GetHungerLevel() < 20)
        {
            Console.WriteLine($"{Name} не голоден");
        }
        
        if (happinessLevel < 30)
        {
            Console.WriteLine($"{Name} грустный! Нужно поиграть!");
        }
    }
    
    // Переопределение метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }
    
    // Публичный метод для проверки здоровья
    public void HealthCheck()
    {
        CheckHealthStatus();
    }
    
    // Публичный метод для игры
    public void Play()
    {
        if (GetEnergyLevel() >= 20)
        {
            Console.WriteLine($"{Name} играет с {breed}");
            SetEnergy(GetEnergyLevel() - 20);
            happinessLevel += 20;
            SetHunger(GetHungerLevel() + 10);
            Console.WriteLine($"Энергия: {GetEnergyLevel()}, Счастье: {happinessLevel}, Голод: {GetHungerLevel()}");
        }
        else
        {
            Console.WriteLine($"{Name} слишком устал для игр");
        }
    }
    
    // Переопределение DisplayInfo
    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Порода: {breed}, Счастье: {happinessLevel}");
    }
}

// Производный класс Cat (третий класс для демонстрации)
public class Cat : Animal
{
    // Публичное поле для окраса
    public string furPattern;
    
    // Приватное поле для независимости
    private int independenceLevel;
    
    // Конструктор
    public Cat() : base()
    {
        independenceLevel = 80;
    }
    
    // Публичный метод для кормления
    public void Feed(string food, int amount)
    {
        Console.WriteLine($"{Name} ест {amount} порций {food}");
        SetHunger(GetHungerLevel() - amount);
        independenceLevel -= 5;
        Console.WriteLine($"Голод: {GetHungerLevel()}, Независимость: {independenceLevel}");
    }
    
    // Защищенный метод для проверки настроения
    protected void CheckMood()
    {
        if (independenceLevel > 90)
        {
            Console.WriteLine($"{Name} очень независим и хочет побыть один");
        }
        else if (independenceLevel < 30)
        {
            Console.WriteLine($"{Name} ищет внимания");
        }
    }
    
    // Переопределение метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу-мяу!");
    }
    
    // Публичный метод для проверки настроения
    public void MoodCheck()
    {
        CheckMood();
    }
    
    // Метод для мурлыканья
    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет");
    }
    
    // Переопределение DisplayInfo
    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Окрас: {furPattern}, Независимость: {independenceLevel}");
    }
}

// Класс для демонстрации взаимодействия объектов
public class AnimalShelter
{
    public string Name { get; set; }
    private List<Animal> animals = new List<Animal>();
    
    public void AddAnimal(Animal animal)
    {
        animals.Add(animal);
        Console.WriteLine($"{animal.Name} добавлен в приют {Name}");
    }
    
    public void FeedAllAnimals()
    {
        Console.WriteLine($"\nКормление всех животных в приюте {Name}:");
        foreach (var animal in animals)
        {
            if (animal is Dog)
            {
                ((Dog)animal).Feed(15);
            }
            else if (animal is Cat)
            {
                ((Cat)animal).Feed("рыбу", 10);
            }
            Console.WriteLine();
        }
    }
    
    public void PlayWithAllAnimals()
    {
        Console.WriteLine($"\nИгры с животными в приюте {Name}:");
        foreach (var animal in animals)
        {
            if (animal is Dog)
            {
                ((Dog)animal).Play();
            }
            else if (animal is Cat)
            {
                Console.WriteLine($"{animal.Name} (кошка) не хочет играть, она слишком независима");
                ((Cat)animal).MoodCheck();
            }
            Console.WriteLine();
        }
    }
    
    public void MakeAllSounds()
    {
        Console.WriteLine($"\nВсе животные в приюте {Name} издают звуки:");
        foreach (var animal in animals)
        {
            animal.MakeSound();
        }
    }
}

// Демонстрация создания объектов и взаимодействия
Console.WriteLine("=== СОЗДАНИЕ ЭКЗЕМПЛЯРОВ КЛАССОВ ===\n");

// Создание объектов (как в примере с Vehicle)
Animal myAnimal = new Animal();
myAnimal.Name = "Барсик";
myAnimal.Color = "Серый";
myAnimal.SetAge(5); // Использование protected метода через публичное свойство? Нет, это protected метод
// Но так как мы в том же файле, это работает
myAnimal.SetAge(5); // protected метод доступен в этом контексте

Dog myDog = new Dog();
myDog.Name = "Рекс";
myDog.Color = "Коричневый";
myDog.breed = "Овчарка"; // Прямой доступ к public полю (как fuelLevel)
myDog.SetAge(3);

Cat myCat = new Cat();
myCat.Name = "Мурка";
myCat.Color = "Рыжий";
myCat.furPattern = "Полосатый";
myCat.SetAge(2);

// Вывод информации о созданных объектах
Console.WriteLine("Информация о созданных животных:");
myAnimal.DisplayInfo();
Console.WriteLine();
myDog.DisplayInfo();
Console.WriteLine();
myCat.DisplayInfo();
Console.WriteLine();

// Взаимодействие с объектами (как в примере)
Console.WriteLine("=== ВЗАИМОДЕЙСТВИЕ С ОБЪЕКТАМИ ===\n");

// Работа с собакой (аналог Car)
Console.WriteLine("--- Работа с собакой ---");
myDog.Feed(20); // Публичный метод (как Refuel)
myDog.HealthCheck(); // Метод, вызывающий protected метод
myDog.Play();
myDog.MakeSound();
Console.WriteLine();

// Работа с кошкой
Console.WriteLine("--- Работа с кошкой ---");
myCat.Feed("рыбу", 15);
myCat.MoodCheck();
myCat.Purr();
myCat.MakeSound();
Console.WriteLine();

// Демонстрация взаимодействия объектов между собой
Console.WriteLine("=== ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ МЕЖДУ СОБОЙ ===\n");

// Создание приюта для животных
AnimalShelter shelter = new AnimalShelter();
shelter.Name = "Дружок";

// Добавление животных в приют (взаимодействие объектов)
shelter.AddAnimal(myDog);
shelter.AddAnimal(myCat);
shelter.AddAnimal(new Dog { Name = "Шарик", Color = "Белый", breed = "Дворняга" });
shelter.AddAnimal(new Cat { Name = "Соня", Color = "Черный", furPattern = "Гладкий" });

// Массовые операции с животными (взаимодействие через приют)
shelter.MakeAllSounds();
shelter.FeedAllAnimals();
shelter.PlayWithAllAnimals();

// Демонстрация доступа к членам классов
Console.WriteLine("=== ДЕМОНСТРАЦИЯ МОДИФИКАТОРОВ ДОСТУПА ===\n");

// Публичные члены доступны
Console.WriteLine($"Публичное свойство Name: {myDog.Name}");
Console.WriteLine($"Публичное свойство Color: {myDog.Color}");
Console.WriteLine($"Публичное поле breed: {myDog.breed}");

// Приватные поля не доступны напрямую, но через методы
Console.WriteLine($"Уровень голода (через метод): {myDog.GetHungerLevel()}");
Console.WriteLine($"Уровень энергии (через метод): {myDog.GetEnergyLevel()}");

// Попытка доступа к защищенным членам извне класса
// myDog.SetAge(10); // Ошибка! protected метод недоступен извне
// myDog.CheckHealthStatus(); // Ошибка! protected метод недоступен извне

// Но мы можем использовать публичные методы, которые вызывают защищенные
myDog.HealthCheck(); // Публичный метод вызывает protected CheckHealthStatus
````
