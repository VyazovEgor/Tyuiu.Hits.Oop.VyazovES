### Структура и Объявление Класса в C#

В C#, класс объявляется с использованием ключевого слова `class`, за которым следует имя класса. Класс может содержать поля (данные), свойства (способ доступа к данным), методы (функциональность) и события. 

#### Свойства

Свойства в C# используются для инкапсуляции данных. Они предоставляют гибкий механизм для чтения, записи или вычисления значений приватных полей. Свойства могут иметь два акцессора: `get` (геттер) для чтения значения и `set` (сеттер) для его изменения.

#### Геттеры и Сеттеры

Геттеры и сеттеры являются частью свойств класса в C#. Геттер (`get`) используется для возврата значения свойства, а сеттер (`set`) — для установки нового значения. Сеттер может включать логику валидации или другие операции перед изменением значения поля.

### Пример Классов Vehicle и Car

Давайте рассмотрим примеры классов `Vehicle` и `Car`, чтобы проиллюстрировать использование свойств, геттеров и сеттеров, а также разницу между статическими и обычными атрибутами и методами.

#### Класс Vehicle

```csharp
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}
```

#### Класс Car

```csharp
public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
```

### Разница между Статическими и Обычными Атрибутами и Методами Класса

- **Статические атрибуты и методы** принадлежат самому классу, а не его экземплярам. Они могут быть вызваны без создания объекта класса. В примере выше `TotalVehicles` является статическим атрибутом, а `DisplayTotalVehicles()` — статическим методом класса `Vehicle`.
- **Обычные атрибуты и методы** принадлежат экземплярам класса и могут иметь разные значения для разных объектов. В примере выше `_color`, `_speed` и `_fuelLevel` являются обычными атрибутами, а методы `Accelerate()` и `Refuel()` — обычными методами.

````
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}


public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
````

````
// Базовый класс Animal (аналог Vehicle)
public class Animal
{
    // Приватные поля (инкапсуляция данных)
    private string _name;
    private int _age;
    private string _species;
    private double _weight;

    // Свойство Name с геттером и сеттером
    public string Name
    {
        get { return _name; }
        set 
        { 
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                throw new ArgumentException("Имя не может быть пустым!");
        }
    }

    // Свойство Age с геттером и сеттером (с валидацией)
    public int Age
    {
        get { return _age; }
        set
        {
            if (value >= 0 && value <= 50) // Животные редко живут больше 50 лет
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст должен быть от 0 до 50 лет!");
        }
    }

    // Свойство Species с геттером и сеттером
    public string Species
    {
        get { return _species; }
        set { _species = value; }
    }

    // Свойство Weight с геттером и сеттером (с валидацией)
    public double Weight
    {
        get { return _weight; }
        set
        {
            if (value >= 0 && value <= 1000) // До 1000 кг
                _weight = value;
            else
                throw new ArgumentOutOfRangeException("Вес должен быть от 0 до 1000 кг!");
        }
    }

    // Статический атрибут для подсчета всех животных
    public static int TotalAnimals { get; private set; } = 0;

    // Конструктор по умолчанию
    public Animal()
    {
        TotalAnimals++;
        _name = "Безымянный";
        _age = 0;
        _species = "Неизвестный вид";
        _weight = 0;
    }

    // Конструктор с параметрами
    public Animal(string name, int age, string species, double weight) : this()
    {
        Name = name;
        Age = age;
        Species = species;
        Weight = weight;
    }

    // Обычный метод - животное ест
    public void Eat(string food, double amount)
    {
        Console.WriteLine($"{Name} ест {food} в количестве {amount} кг");
        Weight += amount * 0.1; // Набирает вес от еды
    }

    // Обычный метод - животное спит
    public void Sleep(int hours)
    {
        Console.WriteLine($"{Name} спит {hours} часов");
    }

    // Виртуальный метод для звуков животного
    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издает звук");
    }

    // Статический метод для отображения общего количества животных
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Общее количество животных: {TotalAnimals}");
    }

    // Метод для получения информации о животном
    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Вид: {Species}, Вес: {Weight} кг";
    }
}

// Производный класс Dog (аналог Car)
public class Dog : Animal
{
    // Приватные поля для специфичных для собак атрибутов
    private string _breed;
    private bool _isTrained;
    private string _favoriteToy;

    // Свойство Breed (порода) с геттером и сеттером
    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    // Свойство IsTrained (дрессирована ли собака)
    public bool IsTrained
    {
        get { return _isTrained; }
        set { _isTrained = value; }
    }

    // Свойство FavoriteToy (любимая игрушка)
    public string FavoriteToy
    {
        get { return _favoriteToy; }
        set { _favoriteToy = value; }
    }

    // Конструктор по умолчанию
    public Dog() : base()
    {
        Species = "Собака";
        Breed = "Дворняга";
        IsTrained = false;
        FavoriteToy = "Мячик";
    }

    // Конструктор с параметрами
    public Dog(string name, int age, double weight, string breed, bool isTrained, string favoriteToy) 
        : base(name, age, "Собака", weight)
    {
        Breed = breed;
        IsTrained = isTrained;
        FavoriteToy = favoriteToy;
    }

    // Переопределение метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    // Специфичный метод для собак
    public void Bark()
    {
        Console.WriteLine($"{Name} громко лает!");
    }

    // Метод для игры
    public void Play()
    {
        Console.WriteLine($"{Name} играет с {FavoriteToy}");
    }

    // Метод для дрессировки
    public void Train()
    {
        if (!IsTrained)
        {
            IsTrained = true;
            Console.WriteLine($"{Name} теперь дрессированная собака!");
        }
        else
        {
            Console.WriteLine($"{Name} уже дрессирована");
        }
    }

    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Порода: {Breed}, Дрессирована: {IsTrained}, Любимая игрушка: {FavoriteToy}";
    }
}

// Производный класс Cat
public class Cat : Animal
{
    // Приватные поля для специфичных для кошек атрибутов
    private string _color;
    private bool _isIndoor;
    private int _livesLeft;

    // Свойство Color (окрас)
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство IsIndoor (домашняя или уличная)
    public bool IsIndoor
    {
        get { return _isIndoor; }
        set { _isIndoor = value; }
    }

    // Свойство LivesLeft (сколько жизней осталось)
    public int LivesLeft
    {
        get { return _livesLeft; }
        set
        {
            if (value >= 1 && value <= 9)
                _livesLeft = value;
            else
                throw new ArgumentOutOfRangeException("У кошек от 1 до 9 жизней!");
        }
    }

    // Конструктор по умолчанию
    public Cat() : base()
    {
        Species = "Кошка";
        Color = "Рыжий";
        IsIndoor = true;
        LivesLeft = 9;
    }

    // Конструктор с параметрами
    public Cat(string name, int age, double weight, string color, bool isIndoor, int livesLeft) 
        : base(name, age, "Кошка", weight)
    {
        Color = color;
        IsIndoor = isIndoor;
        LivesLeft = livesLeft;
    }

    // Переопределение метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу-мяу!");
    }

    // Специфичный метод для кошек
    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет");
    }

    // Метод для охоты
    public void Hunt()
    {
        if (!IsIndoor)
        {
            Console.WriteLine($"{Name} охотится на мышей");
        }
        else
        {
            Console.WriteLine($"{Name} домашняя кошка и не охотится");
        }
    }

    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Окрас: {Color}, Домашняя: {IsIndoor}, Жизней: {LivesLeft}";
    }
}

// Производный класс Bird
public class Bird : Animal
{
    // Приватные поля для специфичных для птиц атрибутов
    private double _wingSpan;
    private bool _canFly;
    private string _featherColor;

    // Свойство WingSpan (размах крыльев)
    public double WingSpan
    {
        get { return _wingSpan; }
        set 
        { 
            if (value > 0)
                _wingSpan = value;
            else
                throw new ArgumentException("Размах крыльев должен быть положительным!");
        }
    }

    // Свойство CanFly (умеет ли летать)
    public bool CanFly
    {
        get { return _canFly; }
        set { _canFly = value; }
    }

    // Свойство FeatherColor (цвет перьев)
    public string FeatherColor
    {
        get { return _featherColor; }
        set { _featherColor = value; }
    }

    // Конструктор по умолчанию
    public Bird() : base()
    {
        Species = "Птица";
        WingSpan = 0.5;
        CanFly = true;
        FeatherColor = "Серый";
    }

    // Конструктор с параметрами
    public Bird(string name, int age, double weight, double wingSpan, bool canFly, string featherColor) 
        : base(name, age, "Птица", weight)
    {
        WingSpan = wingSpan;
        CanFly = canFly;
        FeatherColor = featherColor;
    }

    // Переопределение метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    // Специфичный метод для птиц
    public void Fly()
    {
        if (CanFly)
            Console.WriteLine($"{Name} летит с размахом крыльев {WingSpan} м");
        else
            Console.WriteLine($"{Name} не может летать");
    }

    // Метод для строительства гнезда
    public void BuildNest()
    {
        Console.WriteLine($"{Name} строит гнездо");
    }

    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Размах крыльев: {WingSpan} м, Летает: {CanFly}, Цвет перьев: {FeatherColor}";
    }
}

 Console.WriteLine("=== ДЕМОНСТРАЦИЯ КЛАССОВ ===\n");
        
        // 1. Демонстрация базового класса Animal (как Vehicle)
        Console.WriteLine("--- Базовый класс Animal ---");
        Animal genericAnimal = new Animal("Муся", 2, "Кошка", 4.2);
        Console.WriteLine(genericAnimal.GetInfo());
        genericAnimal.Eat("рыбу", 0.2);
        genericAnimal.Sleep(8);
        genericAnimal.MakeSound();
        Console.WriteLine();
        
        // 2. Демонстрация класса Dog (как Car)
        Console.WriteLine("--- Класс Dog ---");
        Dog dog = new Dog("Рекс", 2, 15.5, "Овчарка", false, "Канат");
        Console.WriteLine(dog.GetInfo());
        dog.MakeSound();
        dog.Bark();
        dog.Play();
        dog.Train();
        Console.WriteLine(dog.GetInfo());
        Console.WriteLine();
        
        // 3. Демонстрация класса Cat
        Console.WriteLine("--- Класс Cat ---");
        Cat cat = new Cat("Мурка", 4, 3.2, "Рыжий", true, 9);
        Console.WriteLine(cat.GetInfo());
        cat.MakeSound();
        cat.Purr();
        cat.Hunt();
        Console.WriteLine();
        
        // 4. Демонстрация класса Bird
        Console.WriteLine("--- Класс Bird ---");
        Bird bird = new Bird("Кеша", 1, 0.5, 0.3, true, "Зеленый");
        Console.WriteLine(bird.GetInfo());
        bird.MakeSound();
        bird.Fly();
        bird.BuildNest();
        Console.WriteLine();
        
        // 5. Демонстрация статического атрибута и метода
        Console.WriteLine("--- Статические члены ---");
        Animal.DisplayTotalAnimals();
        Console.WriteLine($"Всего создано животных: {Animal.TotalAnimals}");
        Console.WriteLine();
        
        // 6. Демонстрация полиморфизма
        Console.WriteLine("--- Полиморфизм ---");
        Animal[] animals = new Animal[]
        {
            new Dog("Шарик", 3, 10, "Такса", true, "Кость"),
            new Cat("Васька", 5, 4, "Черный", false, 7),
            new Bird("Гоша", 2, 0.3, 0.4, true, "Синий"),
            genericAnimal
        };
        
        foreach (var animal in animals)
        {
            Console.WriteLine(animal.GetInfo());
            animal.MakeSound();
            Console.WriteLine();
        }
        
        // 7. Демонстрация валидации в сеттерах
        Console.WriteLine("--- Проверка валидации ---");
        try
        {
            Animal invalidAnimal = new Animal("Тест", 100, "Кошка", 5); // Слишком большой возраст
        }
        catch (ArgumentOutOfRangeException ex)
        {
            Console.WriteLine($"Ошибка валидации: {ex.Message}");
        }
        
        try
        {
            Cat invalidCat = new Cat("Тест", 3, 4, "Серый", true, 15); // Неправильное количество жизней
        }
        catch (ArgumentOutOfRangeException ex)
        {
            Console.WriteLine($"Ошибка валидации у кошки: {ex.Message}");
        }
````
