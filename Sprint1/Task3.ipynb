```
using System;
using System.Collections.Generic;

// Базовый класс Person
public class Person
{
    // Приватные поля для инкапсуляции
    private string _name;
    private int _age;
    private string _gender;
    
    // Свойства с геттерами и сеттерами (с валидацией)
    public string Name
    {
        get { return _name; }
        set 
        { 
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                throw new ArgumentException("Имя не может быть пустым!");
        }
    }
    
    public int Age
    {
        get { return _age; }
        set 
        { 
            if (value >= 0 && value <= 120)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст должен быть от 0 до 120 лет!");
        }
    }
    
    public string Gender
    {
        get { return _gender; }
        set 
        { 
            if (value == "Мужской" || value == "Женский")
                _gender = value;
            else
                throw new ArgumentException("Пол должен быть 'Мужской', 'Женский'");
        }
    }
    
    // Статический атрибут для подсчета всех созданных людей
    public static int TotalPeople { get; private set; } = 0;
    
    // Статический список для хранения всех созданных объектов
    public static List<Person> allPeople = new List<Person>();

    // Конструктор с использованием сеттеров (для валидации)
    public Person(string name, int age, string gender)
    {
        Name = name;
        Age = age;
        Gender = gender;
        TotalPeople++;
        allPeople.Add(this);
    }

    // Виртуальный метод для получения информации о человеке
    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Пол: {Gender}";
    }

    // Виртуальный метод для приветствия
    public virtual void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}.");
    }
    
    // Статический метод для получения общего количества людей
    public static void DisplayTotalPeople()
    {
        Console.WriteLine($"\nВсего создано людей: {TotalPeople}");
    }
    
    // Статический метод для поиска человека по имени
    public static Person FindPerson(string name)
    {
        return allPeople.Find(p => p.Name.Equals(name, StringComparison.OrdinalIgnoreCase));
    }
}

// Производный класс Student (Студент)
public class Student : Person
{
    // Приватные поля
    private string _university;
    private int _course;
    
    // Свойства с геттерами и сеттерами
    public string University
    {
        get { return _university; }
        set 
        { 
            if (!string.IsNullOrWhiteSpace(value))
                _university = value;
            else
                throw new ArgumentException("Название университета не может быть пустым!");
        }
    }
    
    public int Course
    {
        get { return _course; }
        set 
        { 
            if (value >= 1 && value <= 6)
                _course = value;
            else
                throw new ArgumentOutOfRangeException("Курс должен быть от 1 до 6!");
        }
    }
    
    // Список оценок студента
    private List<int> grades = new List<int>();

    // Конструктор с использованием сеттеров
    public Student(string name, int age, string gender, string university, int course) 
        : base(name, age, gender)
    {
        University = university;
        Course = course;
    }

    // Переопределение метода SayHello
    public override void SayHello()
    {
        Console.WriteLine($"Привет! Я студент {Name}, учусь на {Course}-м курсе в {University}.");
    }
    
    // Метод для добавления оценки
    public void AddGrade(int grade)
    {
        if (grade >= 2 && grade <= 5)
        {
            grades.Add(grade);
            Console.WriteLine($"Студенту {Name} добавлена оценка: {grade}");
        }
        else
        {
            Console.WriteLine("Оценка должна быть от 2 до 5!");
        }
    }
    
    // Метод для получения среднего балла
    public double GetAverageGrade()
    {
        if (grades.Count == 0)
            return 0;
        return grades.Average();
    }
    
    // Метод для вывода всех оценок
    public void ShowGrades()
    {
        Console.Write($"Оценки студента {Name}: ");
        if (grades.Count > 0)
        {
            Console.WriteLine(string.Join(", ", grades));
            Console.WriteLine($"Средний балл: {GetAverageGrade():F2}");
        }
        else
        {
            Console.WriteLine("оценок пока нет");
        }
    }
    
    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Университет: {University}, Курс: {Course}, Средний балл: {GetAverageGrade():F2}";
    }

    // Дополнительный метод
    public void Study()
    {
        Console.WriteLine($"{Name} сейчас учится в {University}.");
    }
}

// Производный класс Employee (Работник)
public class Employee : Person
{
    // Приватные поля
    private string _company;
    private decimal _salary;
    
    // Свойства с геттерами и сеттерами
    public string Company
    {
        get { return _company; }
        set 
        { 
            if (!string.IsNullOrWhiteSpace(value))
                _company = value;
            else
                throw new ArgumentException("Название компании не может быть пустым!");
        }
    }
    
    public decimal Salary
    {
        get { return _salary; }
        set 
        { 
            if (value >= 0)
                _salary = value;
            else
                throw new ArgumentOutOfRangeException("Зарплата не может быть отрицательной!");
        }
    }
    
    // Список проектов
    private List<string> projects = new List<string>();

    // Конструктор
    public Employee(string name, int age, string gender, string company, decimal salary) 
        : base(name, age, gender)
    {
        Company = company;
        Salary = salary;
    }
    
    // Метод для добавления проекта
    public void AddProject(string project)
    {
        projects.Add(project);
        Console.WriteLine($"{Name} добавлен в проект: {project}");
    }
    
    // Метод для получения информации о проектах
    public string GetProjects()
    {
        if (projects.Count == 0)
            return "нет проектов";
        return string.Join(", ", projects);
    }
    
    // Метод для повышения зарплаты
    public void RaiseSalary(decimal amount)
    {
        if (amount > 0)
        {
            Salary += amount;
            Console.WriteLine($"Зарплата {Name} повышена до {Salary:C}");
        }
    }
    
    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Зарплата: {Salary:C}, Проекты: {GetProjects()}";
    }

    // Дополнительный метод
    public void Work()
    {
        Console.WriteLine($"{Name} сейчас работает в {Company}.");
    }
}

// Производный класс Teacher (Преподаватель)
public class Teacher : Person
{
    // Приватные поля
    private string _subject;
    private int _experience;
    
    // Свойства с геттерами и сеттерами
    public string Subject
    {
        get { return _subject; }
        set 
        { 
            if (!string.IsNullOrWhiteSpace(value))
                _subject = value;
            else
                throw new ArgumentException("Название предмета не может быть пустым!");
        }
    }
    
    public int Experience
    {
        get { return _experience; }
        set 
        { 
            if (value >= 0 && value <= 70)
                _experience = value;
            else
                throw new ArgumentOutOfRangeException("Стаж должен быть от 0 до 70 лет!");
        }
    }
    
    // Список студентов преподавателя
    private List<Student> students = new List<Student>();

    // Конструктор
    public Teacher(string name, int age, string gender, string subject, int experience) 
        : base(name, age, gender)
    {
        Subject = subject;
        Experience = experience;
    }
    
    // Метод для добавления студента
    public void AddStudent(Student student)
    {
        if (!students.Contains(student))
        {
            students.Add(student);
            Console.WriteLine($"Студент {student.Name} добавлен к преподавателю {Name}");
        }
    }
    
    // Метод для выставления оценки студенту
    public void GradeStudent(Student student, int grade)
    {
        if (students.Contains(student))
        {
            student.AddGrade(grade);
            Console.WriteLine($"Преподаватель {Name} выставил оценку {grade} студенту {student.Name}");
        }
        else
        {
            Console.WriteLine($"Студент {student.Name} не учится у преподавателя {Name}");
        }
    }
    
    // Метод для вывода списка студентов
    public void ShowStudents()
    {
        Console.WriteLine($"Студенты преподавателя {Name} (предмет: {Subject}):");
        if (students.Count > 0)
        {
            foreach (var student in students)
            {
                Console.WriteLine($"  - {student.Name}, курс {student.Course}");
            }
        }
        else
        {
            Console.WriteLine("  Студентов пока нет");
        }
    }
    
    // Переопределение метода SayHello
    public override void SayHello()
    {
        Console.WriteLine($"Здравствуйте! Я преподаватель {Name}, веду предмет '{Subject}'.");
    }

    // Переопределение метода GetInfo
    public override string GetInfo()
    {
        return base.GetInfo() + $", Предмет: {Subject}, Стаж: {Experience} лет, Студентов: {students.Count}";
    }

    // Дополнительный метод
    public void Teach()
    {
        Console.WriteLine($"{Name} преподаёт {Subject}.");
    }
}

// Класс University для демонстрации взаимодействия объектов
public class University
{
    public string Name { get; set; }
    private List<Student> students = new List<Student>();
    private List<Teacher> teachers = new List<Teacher>();
    private List<Employee> employees = new List<Employee>();
    
    public University(string name)
    {
        Name = name;
    }
    
    // Методы для добавления людей в университет
    public void AddStudent(Student student)
    {
        students.Add(student);
        Console.WriteLine($"Студент {student.Name} поступил в университет {Name}");
    }
    
    public void AddTeacher(Teacher teacher)
    {
        teachers.Add(teacher);
        Console.WriteLine($"Преподаватель {teacher.Name} принят на работу в университет {Name}");
    }
    
    public void AddEmployee(Employee employee)
    {
        employees.Add(employee);
        Console.WriteLine($"Сотрудник {employee.Name} принят на работу в университет {Name}");
    }
    
    // Метод для проведения занятий
    public void ConductClasses()
    {
        Console.WriteLine($"\n=== Проведение занятий в университете {Name} ===");
        foreach (var teacher in teachers)
        {
            teacher.Teach();
            if (teacher.GetInfo().Contains("Студентов: 0"))
            {
                Console.WriteLine($"  У преподавателя {teacher.Name} нет студентов!");
            }
        }
    }
    
    // Метод для взаимодействия преподавателей и студентов
    public void OrganizeLearning()
    {
        Console.WriteLine($"\n=== Организация учебного процесса в {Name} ===");
        
        // Распределяем студентов по преподавателям
        if (teachers.Count > 0 && students.Count > 0)
        {
            for (int i = 0; i < students.Count; i++)
            {
                Teacher teacher = teachers[i % teachers.Count];
                teacher.AddStudent(students[i]);
            }
            
            // Преподаватели выставляют оценки
            Console.WriteLine("\n--- Выставление оценок ---");
            foreach (var teacher in teachers)
            {
                foreach (var student in students)
                {
                    // Преподаватель ставит оценку только своим студентам
                    Random rand = new Random();
                    teacher.GradeStudent(student, rand.Next(3, 6));
                }
            }
            
            // Вывод результатов
            Console.WriteLine("\n--- Результаты сессии ---");
            foreach (var student in students)
            {
                student.ShowGrades();
            }
        }
    }
    
    // Метод для выплаты зарплаты
    public void PaySalaries()
    {
        Console.WriteLine($"\n=== Выплата зарплаты в университете {Name} ===");
        foreach (var teacher in teachers)
        {
            Console.WriteLine($"Преподаватель {teacher.Name} получил зарплату");
        }
        foreach (var employee in employees)
        {
            Console.WriteLine($"Сотрудник {employee.Name} получил зарплату {employee.Salary:C}");
        }
    }
    
    // Метод для получения полной информации об университете
    public void GetUniversityInfo()
    {
        Console.WriteLine($"\n=== Информация об университете {Name} ===");
        Console.WriteLine($"Студентов: {students.Count}");
        Console.WriteLine($"Преподавателей: {teachers.Count}");
        Console.WriteLine($"Сотрудников: {employees.Count}");
        Console.WriteLine($"Всего людей: {students.Count + teachers.Count + employees.Count}");
    }
}

// Основная программа
Console.WriteLine("=== ИНДИВИДУАЛЬНЫЙ ПРОЕКТ: СИСТЕМА УПРАВЛЕНИЯ УНИВЕРСИТЕТОМ ===\n");

// Создание объектов с использованием конструкторов
Console.WriteLine("--- СОЗДАНИЕ ОБЪЕКТОВ ---");
Person person = new Person("Иван Петров", 30, "Мужской");
Student student1 = new Student("Анна Смирнова", 20, "Женский", "МГУ", 3);
Student student2 = new Student("Олег Новиков", 22, "Мужской", "МГУ", 4);
Student student3 = new Student("Елена Попова", 19, "Женский", "МГУ", 2);
Employee employee = new Employee("Пётр Сидоров", 35, "Мужской", "ООО Ромашка", 75000);
Teacher teacher1 = new Teacher("Елена Иванова", 45, "Женский", "Математика", 15);
Teacher teacher2 = new Teacher("Дмитрий Соколов", 50, "Мужской", "Физика", 25);

// Вывод информации о созданных объектах
Console.WriteLine("\n--- ИНФОРМАЦИЯ О СОЗДАННЫХ ОБЪЕКТАХ ---");
Console.WriteLine(person.GetInfo());
Console.WriteLine(student1.GetInfo());
Console.WriteLine(student2.GetInfo());
Console.WriteLine(employee.GetInfo());
Console.WriteLine(teacher1.GetInfo());
Console.WriteLine(teacher2.GetInfo());

// Демонстрация статического атрибута
Console.WriteLine("\n--- СТАТИЧЕСКИЕ ЧЛЕНЫ ---");
Person.DisplayTotalPeople();

// Демонстрация работы с оценками
Console.WriteLine("\n--- РАБОТА С ОЦЕНКАМИ ---");
student1.AddGrade(5);
student1.AddGrade(4);
student1.AddGrade(5);
student1.ShowGrades();

student2.AddGrade(3);
student2.AddGrade(4);
student2.ShowGrades();

// Демонстрация работы с проектами сотрудника
Console.WriteLine("\n--- РАБОТА С ПРОЕКТАМИ СОТРУДНИКА ---");
employee.AddProject("Разработка ПО");
employee.AddProject("Тестирование");
employee.RaiseSalary(15000);
Console.WriteLine(employee.GetInfo());

// Демонстрация работы с модификаторами доступа
Console.WriteLine("\n--- ПРОВЕРКА МОДИФИКАТОРОВ ДОСТУПА ---");
try
{
    // Проверка валидации в сеттерах
    Student invalidStudent = new Student("Тест", 200, "Женский", "МГУ", 1);
}
catch (ArgumentOutOfRangeException ex)
{
    Console.WriteLine($"Ошибка валидации возраста: {ex.Message}");
}

try
{
    Student invalidStudent2 = new Student("Тест2", 20, "Неопределенный", "МГУ", 1);
}
catch (ArgumentException ex)
{
    Console.WriteLine($"Ошибка валидации пола: {ex.Message}");
}

// СОЗДАНИЕ УНИВЕРСИТЕТА И ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ
Console.WriteLine("\n=== ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ ===");

// Создание университета
University mgu = new University("МГУ им. Ломоносова");

// Добавление людей в университет
Console.WriteLine("\n--- ДОБАВЛЕНИЕ ЛЮДЕЙ В УНИВЕРСИТЕТ ---");
mgu.AddStudent(student1);
mgu.AddStudent(student2);
mgu.AddStudent(student3);
mgu.AddTeacher(teacher1);
mgu.AddTeacher(teacher2);
mgu.AddEmployee(employee);

// Проведение занятий
mgu.ConductClasses();

// Организация учебного процесса (взаимодействие преподавателей и студентов)
mgu.OrganizeLearning();

// Выплата зарплаты
mgu.PaySalaries();

// Информация об университете
mgu.GetUniversityInfo();

// Демонстрация статического метода поиска
Console.WriteLine("\n--- СТАТИЧЕСКИЙ МЕТОД ПОИСКА ---");
Person foundPerson = Person.FindPerson("Анна Смирнова");
if (foundPerson != null)
{
    Console.WriteLine($"Найден человек: {foundPerson.GetInfo()}");
}

// Финальная демонстрация всех объектов
Console.WriteLine("\n=== ФИНАЛЬНОЕ СОСТОЯНИЕ ВСЕХ ОБЪЕКТОВ ===");
foreach (var p in Person.allPeople)
{
    Console.WriteLine(p.GetInfo());
}
```
