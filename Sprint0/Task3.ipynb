### Вариант задания: №1


<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Person в C#, который будет представлять информацию о
человеке. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Person:
- Атрибуты: Имя (Name), Возраст (Age), Пол (Gender).
- Методы: 
GetInfo(): метод для получения информации о человеке в виде строки.
SayHello(): метод для вывода приветствия от человека.

Требования к производным классам:
1. Студент (Student): Должен содержать дополнительные атрибуты, такие как
Университет (University) и Курс (Course). Метод SayHello() должен быть
переопределен для добавления информации о курсе при приветствии.
2. Работник (Employee): Должен содержать дополнительные атрибуты, такие
как Компания (Company) и Зарплата (Salary). Метод GetInfo() должен быть
переопределен для включения информации о компании и зарплате.
3. Преподаватель (Teacher) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Предмет (Subject) и Стаж работы
(Experience). Метод SayHello() должен быть переопределен для добавления
информации о преподаваемом предмете при приветствии


<h2 style="color:DodgerBlue">Реализация:</h2>

----

```
using System;

// Базовый класс Person
public class Person
{
    // Атрибуты (свойства)
    public string Name { get; set; }
    public int Age { get; set; }
    public string Gender { get; set; }

    // Конструктор
    public Person(string name, int age, string gender)
    {
        Name = name;
        Age = age;
        Gender = gender;
    }

    // Виртуальный метод для получения информации о человеке
    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Пол: {Gender}";
    }

    // Виртуальный метод для приветствия
    public virtual void SayHello()
    {
        Console.WriteLine($"Привет! Меня зовут {Name}.");
    }
}

// Производный класс Student (Студент)
public class Student : Person
{
    // Дополнительные атрибуты
    public string University { get; set; }
    public int Course { get; set; }

    // Конструктор
    public Student(string name, int age, string gender, string university, int course) 
        : base(name, age, gender)
    {
        University = university;
        Course = course;
    }

    // Переопределение метода SayHello для добавления информации о курсе
    public override void SayHello()
    {
        Console.WriteLine($"Привет! Я студент {Name}, учусь на {Course}-м курсе в {University}.");
    }

    // Дополнительный метод для демонстрации
    public void Study()
    {
        Console.WriteLine($"{Name} сейчас учится.");
    }
}

// Производный класс Employee (Работник)
public class Employee : Person
{
    // Дополнительные атрибуты
    public string Company { get; set; }
    public decimal Salary { get; set; }

    // Конструктор
    public Employee(string name, int age, string gender, string company, decimal salary) 
        : base(name, age, gender)
    {
        Company = company;
        Salary = salary;
    }

    // Переопределение метода GetInfo для включения информации о работе
    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Зарплата: {Salary:C}";
    }

    // Дополнительный метод для демонстрации
    public void Work()
    {
        Console.WriteLine($"{Name} сейчас работает в {Company}.");
    }
}

// Производный класс Teacher (Преподаватель) - третий класс
public class Teacher : Person
{
    // Дополнительные атрибуты
    public string Subject { get; set; }
    public int Experience { get; set; } // Стаж в годах

    // Конструктор
    public Teacher(string name, int age, string gender, string subject, int experience) 
        : base(name, age, gender)
    {
        Subject = subject;
        Experience = experience;
    }

    // Переопределение метода SayHello для добавления информации о предмете
    public override void SayHello()
    {
        Console.WriteLine($"Здравствуйте! Я преподаватель {Name}, веду предмет '{Subject}'.");
    }

    // Переопределение метода GetInfo для добавления информации о стаже
    public override string GetInfo()
    {
        return base.GetInfo() + $", Предмет: {Subject}, Стаж: {Experience} лет";
    }

    // Дополнительный метод для демонстрации
    public void Teach()
    {
        Console.WriteLine($"{Name} преподаёт {Subject}.");
    }
}


Console.WriteLine("=== Демонстрация работы классов ===\n");

// Создание объектов разных классов
Person person = new Person("Иван Петров", 30, "Мужской");
Student student = new Student("Анна Смирнова", 20, "Женский", "МГУ", 3);
Employee employee = new Employee("Пётр Сидоров", 35, "Мужской", "ООО Ромашка", 75000);
Teacher teacher = new Teacher("Елена Иванова", 45, "Женский", "Математика", 15);

// Демонстрация работы методов базового класса
Console.WriteLine("--- Базовый класс Person ---");
Console.WriteLine(person.GetInfo());
person.SayHello();

Console.WriteLine("\n--- Класс Student ---");
Console.WriteLine(student.GetInfo()); // Используется метод базового класса
student.SayHello(); // Переопределённый метод
student.Study(); // Собственный метод

Console.WriteLine("\n--- Класс Employee ---");
Console.WriteLine(employee.GetInfo()); // Переопределённый метод
employee.SayHello(); // Используется метод базового класса
employee.Work(); // Собственный метод

Console.WriteLine("\n--- Класс Teacher ---");
Console.WriteLine(teacher.GetInfo()); // Переопределённый метод
teacher.SayHello(); // Переопределённый метод
teacher.Teach(); // Собственный метод

// Демонстрация полиморфизма
Console.WriteLine("\n=== Демонстрация полиморфизма ===");

// Массив объектов базового класса, содержащий объекты производных классов
Person[] people = new Person[]
{
    new Student("Олег Новиков", 22, "Мужской", "СПбГУ", 4),
    new Employee("Мария Кузнецова", 28, "Женский", "Яндекс", 120000),
    new Teacher("Дмитрий Соколов", 50, "Мужской", "Физика", 25),
    person // Объект базового класса
};

// Полиморфный вызов методов
foreach (var p in people)
{
    Console.WriteLine(p.GetInfo()); // Вызывается соответствующий метод для каждого типа
    p.SayHello(); // Вызывается соответствующий метод для каждого типа
    Console.WriteLine(); // Пустая строка для разделения
}
```
